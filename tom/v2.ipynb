```python
# univariate cnn example
# from https://machinelearningmastery.com/how-to-develop-convolutional-neural-network-models-for-time-series-forecasting/

import numpy as np
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
 
import matplotlib as plt
```

```python
# Pick a single time series file
fileName = '/content/gdrive/MyDrive/00000e74ad.npy'
X = np.load(fileName)

print(type(X))
print(X.shape)
print(X.dtype)
print(X.strides)
print(X.data)
print(X)
```

```python
y=np.ones(3)
y[1]=0
print(y.shape)
print(y)

# reshape from [samples, timesteps] into [samples, timesteps, features]
n_steps = 4096
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
print(X.shape)
```

```python
# define model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(n_steps, n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=1000, verbose=0)
```

```python
# demonstrate prediction
fileName = '/content/gdrive/MyDrive/00005bced6.npy'
x_input = np.load(fileName)
print(x_input.shape)
x1 = x_input[0][:]
print(x1.shape)
x1 = x1.reshape((1, n_steps, n_features))
yhat = model.predict(x1, verbose=0)
print(yhat)

#x_input = x_input.reshape((1, n_steps, n_features))
#yhat = model.predict(x_input, verbose=0)
#print(yhat)
```
